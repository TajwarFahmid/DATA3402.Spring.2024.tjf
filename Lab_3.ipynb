*Exercise 2a:*
Write a function that computes the mean of values in a list. Recall the equation for the mean of a random variable $\bf{x}$ computed on a data set of $n$ values $\{ x_i \} = \{x_1, x_2, ..., x_n\}$  is ${\bf\bar{x}} = \frac{1}{n} \sum_i^n x_i$.

````python
# Skeleton
def mean(data):

    # Flatten the nested lists
    flattened_data = [item for sublist in data for item in sublist] if any(isinstance(item, list) for item in data) else data

    if len(flattened_data) == 0:
        return 0  # Return 0 for an empty list to avoid division by zero error

    # Ensure all elements in flattened_data are numeric
    if not all(isinstance(x, (int, float)) for x in flattened_data):
        raise ValueError("All elements in the list must be numerical.")


    total_sum = sum(flattened_data)

    mean_value = total_sum / len(flattened_data)

    return mean_value
````

````python
# Test your solution here
print ("Mean of Data:", mean(data))
````

*Exercise 2b:*
Write a function that computes the variance of values in a list. Recall the equation for the variance of a random variable $\bf{x}$ computed on a data set of $n$ values $\{ x_i \} = \{x_1, x_2, ..., x_n\}$  is ${\bf\langle x \rangle} = \frac{1}{n} \sum_i^n (x_i - {\bf\bar{x}})$.

````python
# Skeleton
def variance(data):


    numeric_data = [x for x in data if isinstance(x, (int, float))]

    if not numeric_data:
        return 0  # Return 0 if there are no numerical values

    mean_value = sum(numeric_data) / len(numeric_data)
    squared_diffs = [(x - mean_value) ** 2 for x in numeric_data]
    variance = sum(squared_diffs) / len(numeric_data)

    return variance






````

## Histogramming

*Exercise 3:* Write a function that bins the data so that you can create a histogram. An example of how to implement histogramming is the following logic:

* User inputs a list of values `x` and optionally `n_bins` which defaults to 10.
* If not supplied, find the minimum and maximum (`x_min`,`x_max`) of the values in x.
* Determine the bin size (`bin_size`) by dividing the range of the function by the number of bins.
* Create an empty list of zeros of size `n_bins`, call it `hist`.
* Loop over the values in `x`
    * Loop over the values in `hist` with index `i`:
        * If x is between `x_min+i*bin_size` and `x_min+(i+1)*bin_size`, increment `hist[i].`
        * For efficiency, try to use continue to goto the next bin and data point.
* Return `hist` and the list corresponding of the bin edges (i.e. of `x_min+i*bin_size`).    

````python
# Solution
def histogram(x, n_bins=10, x_min=None, x_max=None):
    if x_min is None:
        x_min = min(x) if isinstance(x, list) else x
    if x_max is None:
        x_max = max(x) if isinstance(x, list) else x

    # Determine the bin size
    bin_size = (x_max - x_min) / n_bins

    # Create an empty list to store histogram counts
    hist = [0] * n_bins

    # Create a list to store bin edges
    bin_edges = [x_min + i * bin_size for i in range(n_bins)]
    bin_edges.append(x_max)  # Add the last bin edge

    # Loop over the values in x
    for value in x:
        # Check which bin the value belongs to
        for i in range(n_bins):
            if bin_edges[i] <= value < bin_edges[i+1]:
                hist[i] += 1
                break
    return hist, bin_edges
````

Where each line corresponds to a bin and the number of `#`'s are proportional to the value of the data in the bin.

````python
# Solution
def draw_histogram(x, n_bins, x_min=None, x_max=None, character="#", max_character_per_line=20):

    """
    Draw a text-based histogram for the given data.

    Args:
        x (list): A list of numerical values.
        n_bins (int): Number of bins for the histogram.
        x_min (float): Minimum value for binning. If not provided, it will be computed from x.
        x_max (float): Maximum value for binning. If not provided, it will be computed from x.
        character (str): Character to use for representing histogram bars. Defaults to "#".
        max_character_per_line (int): Maximum number of characters per line. Defaults to 20.
    """
    # Define the histogram function
    def histogram(x, n_bins, x_min=None, x_max=None):
        if x_min is None:
            x_min = min(x) if isinstance(x, list) else x
        if x_max is None:
            x_max = max(x) if isinstance(x, list) else x

        # Determine the bin size
        bin_size = (x_max - x_min) / n_bins

        # Create an empty list to store histogram counts
        hist = [0] * n_bins

        # Create a list to store bin edges
        bin_edges = [x_min + i * bin_size for i in range(n_bins)]
        bin_edges.append(x_max)  # Add the last bin edge

        # Loop over the values in x
        for value in x:
            # Check which bin the value belongs to
            for i in range(n_bins):
                if bin_edges[i] <= value < bin_edges[i + 1]:
                    hist[i] += 1
                    break
        return hist, bin_edges

    # Call histogram function to get histogram counts
    hist, bin_edges = histogram(x, n_bins, x_min, x_max)

    # Find the maximum count to scale the histogram
    max_count = max(hist)

    # Determine the scaling factor
    scale_factor = max_count / max_character_per_line

    # Plot the histogram
    for i in range(n_bins):
        bar_length = int(hist[i] / scale_factor)
        print(f"{bin_edges[i]:.1f} - {bin_edges[i + 1]:.1f} | {character * bar_length}")
````
