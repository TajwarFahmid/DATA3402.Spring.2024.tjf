```python
# Question 1
class Matrix:
    def __init__(self, *args):
        if len(args) == 2:
            n, m = args
            self.data = [[0] * m for _ in range(n)]
        elif len(args) == 1 and isinstance(args[0], list):
            self.data = args[0]
            if not all(len(row) == len(self.data[0]) for row in self.data):
                raise ValueError("All rows must have the same number of columns")
        else:
            raise ValueError("Invalid initialization")

    def __getitem__(self, key):
        return self.data[key]

    def __setitem__(self, key, value):
        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value

    def __eq__(self, other):
        if not isinstance(other, Matrix):
            return False
        return self.data == other.data

    def __str__(self):
        return '\n'.join(' '.join(map(str, row)) for row in self.data)


matrix1 = Matrix(3, 4)
print(matrix1)

# Initializing with list of lists
matrix2 = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])
print(matrix2)

# Accessing elements
print(matrix2[1][1])  # Output: 5

# Assigning elements
matrix2[1, 1] = 10
print(matrix2)

# Assigning one matrix to another
matrix3 = Matrix(3, 3)
matrix3[0][0] = 1
matrix3[1][1] = 2
matrix3[2][2] = 3
matrix2 = matrix3
print(matrix2)
```

```python
# Question 2
class Matrix:
    def __init__(self, *args):
        if len(args) == 2:
            n, m = args
            self.data = [[0] * m for _ in range(n)]
        elif len(args) == 1 and isinstance(args[0], list):
            self.data = args[0]
            if not all(len(row) == len(self.data[0]) for row in self.data):
                raise ValueError("All rows must have the same number of columns")
        else:
            raise ValueError("Invalid initialization")

    def __getitem__(self, key):
        if isinstance(key, tuple):
            i, j = key
            if isinstance(i, slice) and isinstance(j, slice):
                return Matrix([row[j] for row in self.data[i]])
            elif isinstance(i, slice):
                return Matrix(self.data[i][j])
            elif isinstance(j, slice):
                return Matrix([[row[j]] for row in self.data[i]])
            else:
                return self.data[i][j]
        else:
            return self.data[key]

    def __setitem__(self, key, value):
        if isinstance(key, tuple):
            i, j = key
            self.data[i][j] = value
        else:
            self.data[key] = value

    def __eq__(self, other):
        if not isinstance(other, Matrix):
            return False
        return self.data == other.data

    def __str__(self):
        return '\n'.join(' '.join(map(str, row)) for row in self.data)

    def shape(self):
        return len(self.data), len(self.data[0])

    def transpose(self):
        return Matrix([list(row) for row in zip(*self.data)])

    def row(self, n):
        return Matrix([self.data[n]])

    def column(self, n):
        return Matrix([[row[n]] for row in self.data])

    def to_list(self):
        return self.data

    def block(self, n_0, n_1, m_0, m_1):
        return Matrix([row[n_0:n_1] for row in self.data[m_0:m_1]])


# Test cases
matrix = Matrix([[1, 2, 3], [4, 5, 6], [7, 8, 9]])

print("Shape:", matrix.shape())
print("Transpose:")
print(matrix.transpose())
print("Row 1:", matrix.row(1))
print("Column 2:", matrix.column(2))
print("Matrix as list of lists:", matrix.to_list())
print("Block (1:3, 0:2):")
print(matrix.block(1, 3, 0, 2))
print("Slicing [1:3, 0:2]:")
print(matrix[1:3, 0:2])
```

```python
# Question 3
def constant(n, m, c):
    """
    Returns a n by m matrix filled with floats of value c.
    """
    return Matrix([[c] * m for _ in range(n)])


def zeros(n, m):
    """
    Returns a n by m matrix filled with floats of value 0.
    """
    return constant(n, m, 0)


def ones(n, m):
    """
    Returns a n by m matrix filled with floats of value 1.
    """
    return constant(n, m, 1)


def eye(n):
    """
    Returns the n by n identity matrix.
    """
    return Matrix([[1 if i == j else 0 for j in range(n)] for i in range(n)])


# Test cases
print("Constant Matrix (3x3) filled with 2:")
print(constant(3, 3, 2))

print("\nZeros Matrix (2x4):")
print(zeros(2, 4))

print("\nOnes Matrix (4x2):")
print(ones(4, 2))

print("\nIdentity Matrix (5x5):")
print(eye(5))
```
