```python
import matplotlib.pyplot as plt
import numpy as np

def create_pair_plots(df, features, signal_col='signal', nrows=3, ncols=3):
    """
    Creates pair plots for a subset of features from the DataFrame.

    Parameters:
    - df: DataFrame containing the features and signal column.
    - features: List of feature names to be plotted.
    - signal_col: Name of the column indicating signal (1) or background (0).
    - nrows, ncols: Number of rows and columns in the plot grid.
    """
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 5*nrows))

    # Flatten the axes array for easy indexing
    axes = axes.flatten()

    for i, ax in enumerate(axes):
        # Calculate row and column index of the upper triangle
        row = i // ncols
        col = i % ncols

        if row < len(features) and col < len(features) and row != col:
            # Plot each class in different colors
            for signal_value, color in zip(df[signal_col].unique(), ['red', 'black']):
                subset = df[df[signal_col] == signal_value]
                ax.scatter(subset[features[col]], subset[features[row]], c=color, label=f"{'Signal' if signal_value else 'Background'}", alpha=0.5)

            ax.set_xlabel(features[col])
            ax.set_ylabel(features[row])
            ax.legend()
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()

# Example usage:
# Selecting a subset of low-level features for demonstration
low_level_features = ['l_1_pT', 'l_1_eta', 'l_1_phi', 'l_2_pT', 'l_2_eta', 'l_2_phi', 'MET', 'MET_phi']
# Assuming 'df' is your DataFrame
# create_pair_plots(df, low_level_features, nrows=2, ncols=2)

# For high-level features, repeat the function call with a different subset of features.
```

```python
import matplotlib.pyplot as plt
import numpy as np

def create_pair_plots_optimized(df, features, signal_col='signal', nrows=3, ncols=3, subsample_size=1000):
    """
    Creates pair plots for a subset of features from the DataFrame with optimizations.

    Parameters:
    - df: DataFrame containing the features and signal column.
    - features: List of feature names to be plotted.
    - signal_col: Name of the column indicating signal (1) or background (0).
    - nrows, ncols: Number of rows and columns in the plot grid.
    - subsample_size: Number of points to sample from each class for plotting.
    """
    fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(5*ncols, 5*nrows))

    # Flatten the axes array for easy indexing
    axes = axes.flatten()

    # Subsample data
    sampled_df = df.groupby(signal_col).apply(lambda x: x.sample(n=subsample_size, replace=True) if len(x) > subsample_size else x).reset_index(drop=True)

    for i, ax in enumerate(axes):
        row = i // ncols
        col = i % ncols

        if row < len(features) and col < len(features) and row != col:
            for signal_value, color in zip(sampled_df[signal_col].unique(), ['red', 'black']):
                subset = sampled_df[sampled_df[signal_col] == signal_value]
                ax.scatter(subset[features[col]], subset[features[row]], c=color, label=f"{'Signal' if signal_value else 'Background'}", alpha=0.5)

            ax.set_xlabel(features[col])
            ax.set_ylabel(features[row])
            ax.legend()
        else:
            ax.axis('off')

    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import pandas as pd
from tabulate import tabulate

def display_cov_corr_matrices(df, features, sig_figs=2):
    """
    Computes and displays the covariance and correlation matrices for the specified features in the given DataFrame.

    Parameters:
    - df: pandas DataFrame containing the dataset.
    - features: List of feature names for which to compute the matrices.
    - sig_figs: Number of significant figures to display in the matrices.
    """
    # Compute the covariance and correlation matrices
    cov_matrix = np.cov(df[features].T)
    corr_matrix = np.corrcoef(df[features].T)

    # Format the matrices for display
    formatted_cov_matrix = [[f"{item:.{sig_figs}f}" for item in row] for row in cov_matrix]
    formatted_corr_matrix = [[f"{item:.{sig_figs}f}" for item in row] for row in corr_matrix]

    # Generate headings
    headings = features

    # Create tabulated versions of the matrices
    tabulated_cov_matrix = tabulate(formatted_cov_matrix, headers=headings, tablefmt="grid")
    tabulated_corr_matrix = tabulate(formatted_corr_matrix, headers=headings, tablefmt="grid")

    # Display the matrices
    print("Covariance Matrix:")
    print(tabulated_cov_matrix)
    print("\nCorrelation Matrix:")
    print(tabulated_corr_matrix)

# Example usage:
# Assuming 'df' is your DataFrame and 'selected_features' is a list of features you're interested in
# selected_features = ['lepton 1 pT', 'lepton 1 eta', 'missing energy magnitude']
# display_cov_corr_matrices(df, selected_features)
```

```python
def plot_tpr_fpr(df, feature, signal_col='signal'):
    """
    Plots TPR and FPR for the given feature across a range of cutoff values (x_c).

    Parameters:
    - df: DataFrame containing the data.
    - feature: The feature (observable) to analyze.
    - signal_col: The column name indicating signal (1) or background (0).
    """
    x_c_values = np.linspace(df[feature].min(), df[feature].max(), 100)
    tpr_values = []
    fpr_values = []

    for x_c in x_c_values:
        # Applying the criterion x > x_c for demonstration
        true_positives = df[(df[signal_col] == 1) & (df[feature] > x_c)].shape[0]
        false_positives = df[(df[signal_col] == 0) & (df[feature] > x_c)].shape[0]
        total_signal = df[df[signal_col] == 1].shape[0]
        total_background = df[df[signal_col] == 0].shape[0]

        tpr = true_positives / total_signal if total_signal else 0
        fpr = false_positives / total_background if total_background else 0

        tpr_values.append(tpr)
        fpr_values.append(fpr)

    # Plotting
    plt.figure(figsize=(10, 5))
    plt.plot(x_c_values, tpr_values, label='TPR (Signal Efficiency)', color='blue')
    plt.plot(x_c_values, fpr_values, label='FPR (Background Efficiency)', color='red')
    plt.xlabel(f'Cutoff Value ({feature})')
    plt.ylabel('Efficiency')
    plt.title(f'TPR and FPR as Function of {'l_1_phi'}')
    plt.legend()
    plt.grid(True)
    plt.show()
```
